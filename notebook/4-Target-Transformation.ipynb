# Load Dependency

```python
from script.TargetTransformation import box_cox_trans,ee2ddG,log_trans,vis_transformed_result
from script.DescriptorProcess import maxminscale
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load Data

```python
dataset = pd.read_csv('../dataset.csv',index_col=0)
ee = dataset['ee'].to_numpy()
T = dataset['Temperature'].to_numpy()+273.15
```

# Perform Target Transformation

## Original Distribution

```python
algorithm = 'No'
vis_transformed_result(ee,algorithm,color='royalblue')
```

## Eyring-Polanyi Equation

```python
transformed_target = maxminscale(ee2ddG(ee,T))
algorithm = 'Eyring-Polanyi Equation'
vis_transformed_result(transformed_target,algorithm,color='forestgreen')
```

## Box-Cox

```python
transformed_target = maxminscale(box_cox_trans(ee,15))
algorithm = 'Box-Cox Transformation'
vis_transformed_result(transformed_target,algorithm,color='coral')
```

## Logarithmic

```python
transformed_target = maxminscale(log_trans(ee))
algorithm = 'Logarithmic Transformation'
vis_transformed_result(transformed_target,algorithm,color='mediumturquoise')
```
